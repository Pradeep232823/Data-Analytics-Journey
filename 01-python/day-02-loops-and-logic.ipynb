## Objective
Understand how loops and control flow work in Python and apply them to solve basic logical problems.


## for Loop (Basics)

```python
# Print numbers from 1 to 10 using for loop

for i in range(1, 11):
    print(i)
```

## while Loop

```python
# Print numbers from 1 to 5 using while loop

i = 1
while i <= 5:
    print(i)
    i += 1
```

## break and continue

```python
# Demonstrate break and continue

for i in range(1, 11):
    if i % 4 == 0:
        continue
    if i == 9:
        break
    print(i)
```

## Sum of Numbers (1 to n)

```python
# Find sum of numbers from 1 to n

n = int(input("Enter a number: "))
total = 0

for i in range(1, n + 1):
    total += i

print("Sum:", total)
```

## Even / Odd Check in a List

```python
# Check even and odd numbers in a list

numbers = [2, 5, 8, 11, 14]

for num in numbers:
    if num % 2 == 0:
        print(num, "is Even")
    else:
        print(num, "is Odd")
```

## Count Vowels in a String

```python
# Count number of vowels in a string

text = input("Enter a word: ")
count = 0

for char in text.lower():
    if char in "aeiou":
        count += 1

print("Number of vowels:", count)
```

## Multiplication Table

```python
# Print multiplication table of a number

num = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

## Key Takeaways
- Loops help automate repetitive tasks
- for loops work well with ranges and sequences
- while loops are useful when repetition depends on a condition
- Clean logic is more important than complex syntax
