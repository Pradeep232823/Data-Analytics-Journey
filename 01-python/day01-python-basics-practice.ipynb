# Order of operations

The arithmetic we learned in primary school has conventions about the order in which operations are evaluated. Some remember these by a mnemonic such as 
# PEMDAS - Parentheses, Exponents, Multiplication/Division, Addition/Subtraction.

Python follows similar rules about which calculations to perform first. They're mostly pretty intuitive.

```python
8 - 3 + 2
```

```python
-3 + 4 * 2
```

```python
hat_height_cm = 25
my_height_cm = 190
# How tall am I, in meters, when wearing my hat?
total_height_meters = hat_height_cm + my_height_cm / 100 #Here operations done based on PEMDAS
print("Height in meters =", total_height_meters, "?") #Wrong value prints
```

```python
#Parentheses are useful here. You can add them to force Python to evaluate sub-expressions in whatever order you want.

total_height_meters = (hat_height_cm + my_height_cm) / 100 #Paranthesis helps to separate the expressions as we need
print("Height in meters =", total_height_meters)
```

# Defining functions
Builtin functions are great, but we can only get so far with them before we need to start defining our own functions. Below is a simple example.

```python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    
    >>> least_difference(1, 5, -5)
    4"""
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
```

This creates a function called least_difference, which takes three arguments, a, b, and c.

Functions start with a header introduced by the def keyword. The indented block of code following the : is run when the function is called.

return is another keyword uniquely associated with functions. When Python encounters a return statement, it exits the function immediately, and passes the value on the right hand side to the calling context.

```python
print(
    least_difference(1, 10, 100),
    least_difference(1, 10, 10),
    least_difference(5, 6, 7), # Python allows trailing commas in argument lists. How nice is that?
)
```

```python
help(least_difference)
```

# Functions that don't return

What would happen if we didn't include the return keyword in our function?

```python
def least_difference(a, b, c):
    """Return the smallest difference between any two numbers
    among a, b and c.
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    min(diff1, diff2, diff3)
    
print(
    least_difference(1, 10, 100),
    least_difference(1, 10, 10),
    least_difference(5, 6, 7),
)
```

**Without the return statement the function call returns value None**

## 3. Average of Even Numbers Only

### Task

1. From a list of numbers
2. Calculate the average of only even numbers

**Hint**
1. Use a loop or list comprehension
2. Check even using % 2 == 0

```python
lst=[1,2,3,4,5,6]
even_sum=0
even_count=0
for i in lst:
    if i % 2 == 0:
        even_sum += i
        even_count += 1
print(f"Average of even numbers in list is {even_sum/even_count}")
```

## 5. Count Numbers Above Average
### Task
1. Calculate the average
2. Count how many numbers are greater than the average in a list

```python
lst=[26,45,20,42,34,16]
avg=sum(lst)/len(lst)
count=0
for i in lst:
    if i>avg:
        count+=1
print("Average:",avg)
print("Count of numbers above average:",count)
```
