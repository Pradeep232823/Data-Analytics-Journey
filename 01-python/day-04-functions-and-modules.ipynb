## Objective

## Basic Function

```python
# Function to calculate sum and average

def calculate_sum_avg(numbers):
    total = sum(numbers)
    avg = total / len(numbers)
    return total, avg

nums = [10, 20, 30, 40]
print(calculate_sum_avg(nums))
```

## Even or Odd Function

```python
# Function to check even or odd

def check_even_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

print(check_even_odd(15))
```

## Area Calculation Function

```python
# Function to calculate area

def calculate_area(shape, a, b=0):
    if shape == "square":
        return a * a
    elif shape == "rectangle":
        return a * b

print(calculate_area("square", 4))
print(calculate_area("rectangle", 4, 6))
```

## Function with Return Value

```python
# Function returning a value

def greet(name):
    return f"Hello {name}"

print(greet("Pradeep"))
```

## Lambda Function

```python
# Lambda function example

square = lambda x: x * x
print(square(5))
```

## random Module

```python
# Using random module

import random

print("Random number:", random.randint(1, 10))
print("Random choice:", random.choice(["Apple", "Banana", "Mango"]))
```

## Key Takeaways
- Functions help write reusable and clean code
- return allows functions to produce output
- Lambda functions are useful for small operations
- Python modules provide ready-to-use functionality
