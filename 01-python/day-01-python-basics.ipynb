# Day 1 – Python Basics

### Variables & Data Types

```python
name = "Pradeep"
age = 21
height = 5.9
is_learning_da = True

print(name, type(name))
print(age, type(age))
print(height, type(height))
print(is_learning_da, type(is_learning_da))
```

### User Input & Type Casting

```python
age = int(input("Enter your age: "))
print(f"Next year you will be {age + 1} years old")
```

### Operators

```python
a, b = 10, 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

print("Greater than:", a > b)
print("Equal:", a == b)
```

### Conditional Statements

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### Mini Problem (Real‑World Logic)

```python
temp = int(input("Enter temperature: "))

if temp > 30:
    print("Hot")
elif temp >= 15:
    print("Moderate")
else:
    print("Cold")
```

## Key Takeaways
- Python basics are essential for data cleaning and logic building
- Conditional statements help in decision-based analysis
- Writing clean and readable code matters more than solving many problems
