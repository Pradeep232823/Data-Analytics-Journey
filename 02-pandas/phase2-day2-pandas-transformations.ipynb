# Pandas Transformations

## Objective
Apply business logic and transformations to derive meaningful insights from clean data.

```python
import pandas as pd
import numpy as np

df=pd.read_csv("data/superstore_cleaned.csv",encoding="latin1")
```

```python
df_high_sales = df[df["sales"]>5000]
df_high_sales
```

```python
High_sale_product = df[df["sales"] == df["sales"].max()]
High_sale_product
```

```python
High_Tech_Sales = df[(df["category"] == "Technology") & (df["sales"]>10000)]
High_Tech_Sales
```

```python
High_Office_Furn_Sales = df[((df["category"] == "Office Supplies") | (df["category"] == "Furniture")) & (df["sales"]>5000)]
High_Office_Furn_Sales
```

Filtering was used to identify relevant records for various product categories.
This allows for the focused analysis of highly successful sales segments and enables meaningful category-level insights.

```python
def profit(sales):
    if sales>9000:
        return "High Profit"
    elif sales>3000:
        return "Medium Profit"
    else:
        return "Low Profit"

df["profit"] = df["sales"].apply(profit)
df[df["profit"]=="Medium Profit"]
```

New columns were created to translate raw sales values into business-relevant metrics.
These derived fields help segment performance, simplify comparisons, and support deeper analytical insights.

```python
highest_sales_subcategory = df.loc[
    df.groupby("category")["sales"].idxmax()
]

highest_sales_subcategory[
    ["category", "sub-category", "sales"]
]
```

```python
df["high_sales_flag"] = df["sales"].apply(
    lambda x: "Yes" if x >= 10000 else "No"
)
```

Business rules were applied to classify records and support decisions.
These rules transform raw data into structured insights, thus enabling segmentation.
It consists of a set of mechanisms related to prioritization and performance evaluation, which must be in line with business objectives.

```python
df.sort_values("sales",ascending=False)
```

```python
df.sort_values("category")
```

"Sorting" was used to arrange the data according to certain key metrics and respective categories.
This makes it easier to highlight top- and bottom-performing records, which can help in understanding trends, etc.

```python
df.to_csv(
    "data/superstore_transformations.csv",
    index=False
)
```

## Key Takeaways
- Business-focused filtering was applied to isolate relevant records and analyze high-performing segments.
- New derived columns were created to translate raw sales data into actionable business metrics.
- Business rules and sorting logic were used to organize data and highlight key trends.
- The transformed dataset is now structured to support deeper analysis, reporting, and visualization.
