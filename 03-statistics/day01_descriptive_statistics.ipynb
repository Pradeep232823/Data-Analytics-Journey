This notebook applies mean, median, and mode to small datasets and records insights from the results.


## Concepts (Quick Reference)
- Mean: average value, sensitive to outliers
- Median: middle value, robust to outliers
- Mode: most frequent value

```python
import statistics

data_normal = [10, 12, 14, 16, 18]
data_outlier = [10, 12, 14, 100]

print(statistics.mean(data_normal), statistics.median(data_normal))
print(statistics.mean(data_outlier), statistics.median(data_outlier))
```

## Insights
- In the dataset with an outlier, the mean increased significantly.
- The median remained stable and better represented the typical value.
- Median is preferred when data contains extreme values.
